# Loading Country data file

# Merging the two datasets

```python
final_df = pd.merge(df, df_country, on = 'Country Code', how='left')
final_df.head()
```

# Geographic Transaction Analysis

```python
country_names = final_df['Country'].value_counts().index
country_names
```

```python
#Top 3 countries that uses Zomato
plt.pie(country_values[:3], labels= country_names[:3], autopct='%1.2f%%' )
```

# Customer Rating Analysis

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Obersvations
1. When rating is between 4.5 to 4.9, it indicates Excellent service
2. When rating is between 4.0 to 4.4, it indicates Very Good service
3. When rating is between 3.5 to 3.9, it indicates Good service
4. When rating is between 2.5 to 3.4, it indicates Average service
5. When rating is between 1.8 to 2.4, it indicates Poor service
6. When rating is 0, it indicates customers have not rated the restaurant.





```python
ratings.head()
```

```python
plt.figure(figsize=(12,6))
ax = sns.barplot(x='Aggregate rating',y='Rating Count', data=ratings, hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
plt.show()
```

Observations:

    1. Not rated rating is too high
    2. Maximum number of ratings is in between 2.5 and 3.4

```python
# count plot - how frequent each of the rating color appears in the dataframe
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
